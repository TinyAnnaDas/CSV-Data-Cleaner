Sorting


```python
import pandas as pd
```

```python
people = {
    "first": ["Corey", "Jane", "John", "Adam"],
     "last": ["Shafer", "Doe", "Doe", "Doe"],
    "email": ["CoreyMShafer@gmail.com", "JaneDoe@gmaio.com", "JohnDoe@gmail.com", "A@gmail.com"]
}
```

```python
df = pd.DataFrame(people)
```

```python
# sort this dataframe by last name
df.sort_values(by="last")
```

```python
df.sort_values(by="last", ascending=False)
```

```python
# sort on multiple columns
# sort one column, if duplicate values in that column then sort by another related column
df.sort_values(by=["last", "first"], ascending=False)
# Here the last name has duplicate values Doe, so it then sort by first
```

```python
# one column in descending order and the other in ascending order
df.sort_values(by=["last", "first"], ascending=[False, True], inplace=True)
```

```python
df.sort_index()
```

```python
# if you want to sort a single column
df["last"].sort_values()
```
