#### Grouping of Data

```python
# If we want to include other columns
# group by operation - Split, Apply Function - Combine Results
df['Country'].value_counts()
```

```python
# Splitting the object
country_group = df.groupby(["Country"])
country_group
```

```python
country_group.get_group('India') # the result is same as doing a filter
```

```python
filt = df['Country'] == "India" #
df.loc[filt]["Age"].value_counts() # it gives teh result for just one country.
```

```python
filt = df["Country"] == "India"
df.loc[filt]["LanguageHaveWorkedWith"].str.contains("Python").sum()
```

```python
country_group['LanguageHaveWorkedWith'].apply(lambda x: x.str.contains("Python").sum())
```

```python
# What % of people from each country know Python ?
country_respondence = df["Country"].value_counts()
country_respondence
country_uses_python = country_group['LanguageHaveWorkedWith'].apply(lambda x: x.str.contains("Python").sum())
country_uses_python
```

```python
# Combine the above two country_respondence and country_uses_python

python_df = pd.concat([country_respondence, country_uses_python], axis='columns', sort=False)
python_df
```

```python
python_df.rename(columns={"count": "NumRespondents", "LanguageHaveWorkedWith": "NumKnowsPython"}, inplace=True)
```

```python
python_df
```

```python
python_df['PctKnowsPython'] = (python_df["NumKnowsPython"] /python_df["NumRespondents"]) * 100
python_df
```

```python
# sort the above dataframe, countries by largest percentage who knows python
python_df.sort_values(by="PctKnowsPython", ascending=False, inplace=True)
```
